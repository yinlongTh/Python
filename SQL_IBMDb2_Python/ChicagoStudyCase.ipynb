```python
%%sql
SELECT *
FROM (SELECT CC.community_area_number, count(CC.id) AS Number_of_Crime, CS.community_area_name
FROM CHICAGOCRIME CC, CHICAGOCENSUS CS
WHERE CC.community_area_number = CS.community_area_number
GROUP BY CC.community_area_number, CS.community_area_name
ORDER BY -Number_of_Crime) AS Fdt
LIMIT 1
```

Ex1 : Using Joins

```python
%%sql
select CS.community_area_name, SC.name_of_school, CS.hardship_index
from CHICAGOSCHOOLS AS SC 
RIGHT OUTER JOIN (SELECT *
      FROM CHICAGOCENSUS
      WHERE hardship_index =98) AS CS ON SC.community_area_number = CS.community_area_number;
```

```python
%%sql
select CC.case_number, CC.primary_type, CS.community_area_name
from (SELECT community_area_number, case_number, primary_type
FROM CHICAGOCRIME
WHERE location_description like '%SCHOOL%') AS CC, CHICAGOCENSUS AS CS
```
