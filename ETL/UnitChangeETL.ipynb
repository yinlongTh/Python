```python
import glob                        
import pandas as pd               
from datetime import datetime

import urllib.request
import zipfile
```

```python
#Files download and unzip
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0221EN-SkillsNetwork/labs/module%206/Lab%20-%20Extract%20Transform%20Load/data/source.zip'
#This data source is released under the terms of the MIT License by Copyright Â© 2020 IBM Corporation.
file_name = 'source.zip'
urllib.request.urlretrieve(url, file_name)
with zipfile.ZipFile('./' + file_name, 'r') as zip_ref:
    zip_ref.extractall('./')
```

```python
#Extract function
##CSV
def CSVExt(file): 
    dtFrame = pd.read_csv(file)
    return dtFrame

##JSON
def JSONExt(file):
    dtFrame = pd.read_json(file,lines=True)
    return dtFrame
    
    

#Extract loop function
def extLoop():
    dtFrame = pd.DataFrame(columns=["name","height","weight"])
    
    #CSV loop
    for x in glob.glob("*.csv"):
        dtFrame = dtFrame.append(CSVExt(x), ignore_index = True)
        
    #JSON loop
    for x in glob.glob("*.json"):
        dtFrame = dtFrame.append(JSONExt(x), ignore_index = True)
        
        
   
    return dtFrame 
```

```python
#Transform : inche to m, pound to kg
def convt(data):
    data['height'] = data.height*0.0254
    data['weight'] = data.weight * 0.45359237
    return data
```

```python
readData = extLoop()
tranData = convt(readData)

display(tranData)
tranData.to_csv('isUnitData.csv', index=False)
```
