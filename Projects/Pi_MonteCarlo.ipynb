### Calculating $\pi$ : Monte Carlo Simulation

For a square-perfectly-fitted circle 
$\pi \approx 4\cdot N_{circle}/N_{square}$.

```python
import random
import math
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

def PiApprox(power):

 N_circle = 0 
 N_square = int(math.pow(10,power))

 R = 1 #Circle Radius

 for i in range(N_square):
    x = random.random()
    y = random.random()
    if x**2 + y**2 <= R:
        N_circle = N_circle + 1 
    
 Pi_val = 4.0 * N_circle / N_square

 return Pi_val
```

##### How many sample start giving us an acceptable error  

```python
x = np.array([1,2,3,4,5,6,7], dtype='int8')
length = len(x)
y = np.zeros(length)
index = 0

for j in x :
  y[index] = PiApprox(j)
  index += 1

error = ((math.pi-y)/math.pi)*100

dtFrame = pd.DataFrame(columns=["Power","Error (%)"])
dtFrame['Power'] = x
dtFrame['Error (%)'] = error

display(dtFrame)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,5,12])
ax.plot(x,error) 
ax.set_title('Calculating $\pi$ : Monte Carlo Simulation')
ax.set_xlabel('Numper of Sample (10^x)')
ax.set_ylabel('Error (%)')
```
