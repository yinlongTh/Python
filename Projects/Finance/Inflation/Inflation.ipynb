```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Calculate Future Value
def fv(inflation,cost,year):
    fv_item = np.zeros(year)
    fv_item[0] = cost

    for i in range(1,year,1):
        fv_item[i] = fv_item[i-1]*(1+inflation/100)  

    return fv_item
```

```python
#Parameter Setting
interest = 4.0 #%
inflation = 5.0 #%

cost = 100000
year = 20 #years

#Function calls
fv_item = fv(inflation,cost,year)
fv_saving = fv(interest,cost,year)

real_fv_saving = (fv_saving - fv_item) + cost
```

```python
#Visualization
plt.plot(range(1,year+1,1), fv_item, label = "Item Cost")
plt.plot(range(1,year+1,1), fv_saving, label = "Your Saving")
plt.plot(range(1,year+1,1), real_fv_saving, label = "Real Saving Value", linestyle="-.")
plt.ylabel("Value (USD)")
plt.xlabel("Year")
plt.legend()
plt.show()

print("In " + str(year) + " years, your saving becomes : $ " + str(round(fv_saving[year-1],2)))
print("Because of the inflation, item will cost : $ " + str(round(fv_item[year-1],2)))
```
